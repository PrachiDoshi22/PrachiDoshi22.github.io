```python
ufos = pd.read_csv('https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_AOUAOG_fall2021/master/week05/data/ufo-scrubbed-geocoded-time-standardized-00.csv', 
                  names=["date", "city", "state", "country", 
                        "shape", "duration_seconds", "duration", 
                        "comment", "report_date", "latitude", "longitude"], 
                  parse_dates = ["date", "report_date"]) # tell pandas that these columns are dates
```

```python
ufos.head()
```

```python
ufos = ufos.fillna({'country':'unknown', 'state': 'unknown', 'shape':'unknown'})
ufos.head(10)
```

```python
nsamples = 10000
downSampleMask = np.random.choice(range(len(ufos)-1), nsamples, replace=False)
ufos_sample = ufos.loc[downSampleMask]
```

```python
style = {'description_width': 'auto'}

# Creating dropdown list for the column names for X axis. 
# Longitude is set as default.
X_axis_field = widgets.Dropdown(
    options=["date", "state", "country", "report_date", "longitude", "latitude"],
    value="longitude", description='Field on X axis', style = style
)

# Creating dropdown list for the column names for Y axis. 
# Latitude is set as default.
Y_axis_field = widgets.Dropdown(
    options=["date", "shape", "report_date", "latitude", "longitude"],
    value="latitude", description='Field on Y axis', style = style
)


# creating dropdwon list for the plot style. 
# seaborn is set as default.
plot_style = widgets.Dropdown(
    options=plt.style.available, value="seaborn", 
    description="Plot Style", style = style
)
```

```python
# Creating a colormap.
colormap_options = widgets.Dropdown(
    options=plt.colormaps(), value="RdBu", 
    description="Choice of colormap", style = style
)

log_toggle = widgets.Checkbox(value = True, description='Logarithmic Duration in seconds:')
```

```python
# Getting all the widgets together for the visualisation.
wid_for_scatter = widgets.VBox([widgets.HBox([X_axis_field, Y_axis_field]),
                    widgets.HBox([plot_style, colormap_options]),
                    widgets.HBox([log_toggle])])

# Function to create a scatter plot.
def make_scatter_plot(x, y,style, colormap, log):
    with plt.style.context(style):
        fig, ax = plt.subplots(figsize=(30,15))
        if log:
            c_map_values = np.log10(ufos_sample['duration_seconds'])
        else:
            c_map_values = ufos_sample['duration_seconds']
        scat_plt = ax.scatter(ufos_sample[x], ufos_sample[y], c = c_map_values, 
                   cmap = colormap)
        ax.set_xlabel(x, fontsize = 25)
        ax.set_ylabel(y, fontsize = 25)
        ax.set_title(x+" vs "+y,fontsize = 30)
        plt.colorbar(scat_plt, extend = 'both', label = 'Duration in Seconds(Log Scale)')
        plt.show()

# Connecting the plot with the interactive widgets.
plt_out = widgets.interactive_output(make_scatter_plot, 
                                 {'style': plot_style, 'colormap': colormap_options, 
                                  'x': X_axis_field, 'y': Y_axis_field,
                                 'log': log_toggle})

# Displaying the plot with the widgets.
display(wid_for_scatter, plt_out)
```
